```python
import matplotlib.pyplot as plt
import numpy as np

species = ("100", "200", "500")
penguin_means = {
    'EAPTS': (38.35, 58.43, 44.98),
    'GAI-Edge-Cloud': (68.79, 88.83, 97.50),
    'G-Edge-Cloud': (189.95, 195.82, 217.19),
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('CO2')
ax.set_xlabel('Number of Tasks of Transport Workloads')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(4, 298)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
species = ("100", "200", "500")
penguin_means = {
    'EAPTS': (6, 9, 11.98),
    'GAI-Edge-Cloud': (11, 17, 15),
    'G-Edge-Cloud': (18, 21, 19),
}

# Define colors for each dataset
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  # Blue, Orange, Green

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

# Create plot
fig, ax = plt.subplots(layout='constrained')

# Plot bars with different colors
for (attribute, measurement), color in zip(penguin_means.items(), colors):
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute, color=color)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Set labels, title, and custom x-axis tick labels
ax.set_ylabel('Total Execution Time (Minutes)')
ax.set_xlabel('Number of Tasks of Transport Workloads')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 30)

# Add horizontal grid lines
ax.grid(axis='y', linestyle='--', linewidth=0.5, color='gray')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Apply MATLAB-like classic style
plt.style.use('classic')

# Data
species = ("100", "200", "500")
penguin_means = {
    'EAPTS': (4, 5, 9),
    'GAI-Edge-Cloud': (7, 9, 15),
    'G-Edge-Cloud': (11, 15, 28),
}

# Define colors for each dataset
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  # Blue, Orange, Green

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

# Create plot
fig, ax = plt.subplots(figsize=(8, 6))

# Plot bars with different colors
for (attribute, measurement), color in zip(penguin_means.items(), colors):
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute, color=color)
    ax.bar_label(rects, padding=3, fontsize=10, color='black')
    multiplier += 1

# Set labels, title, and custom x-axis tick labels
ax.set_ylabel('IoT Time (Minutes)', fontsize=12, weight='bold')
ax.set_xlabel('Number of Video Transport Data', fontsize=12, weight='bold')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3, fontsize=10)
ax.set_ylim(0, 30)

# Add horizontal grid lines
ax.grid(axis='y', linestyle='--', linewidth=0.5, color='gray')

# Make tick labels in a similar style to MATLAB
ax.tick_params(axis='both', which='major', labelsize=10)

plt.show()
```
